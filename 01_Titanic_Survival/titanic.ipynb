```python
import pandas as pd
import numpy as np
```

```python
train_df = pd.read_csv("train.csv") #verileri yükledim.
test_df = pd.read_csv("test.csv")
```

```python
train_df
```

```python
test_df
```

```python
train_df = train_df.drop(["Name","Ticket","Cabin"], axis=1) #gereksiz sütunları kaldırdım.
test_df = test_df.drop(["Name","Ticket","Cabin"], axis=1) #hem train hem test sütunu için de.
```

```python
train_df
```

```python
test_df
```

```python
from sklearn.impute import SimpleImputer #simleımputer ile eksik verileri doldurdum.
imputer_sayisal = SimpleImputer(missing_values=np.nan, strategy="mean") #sayısal veri için.
train_df[["Age","Fare"]] = imputer_sayisal.fit_transform(train_df[["Age","Fare"]])
test_df[["Age","Fare"]] = imputer_sayisal.transform(test_df[["Age","Fare"]])

imputer_kategorik = SimpleImputer(missing_values=np.nan, strategy="most_frequent") #kategorik veriler için en çok tekrar eden
train_df[["Embarked"]] = imputer_kategorik.fit_transform(train_df[["Embarked"]])
test_df[["Embarked"]] = imputer_kategorik.transform(test_df[["Embarked"]])
```

```python
from sklearn.preprocessing import LabelEncoder, OneHotEncoder
le = LabelEncoder()

train_df["Sex"] = le.fit_transform(train_df["Sex"]) #cinsiyetleri 0 1 e çevirdim
test_df["Sex"] = le.transform(test_df["Sex"])
train_df
test_df
```

```python
#-------Train Verisi İçin İşlemler-----------
ohe = OneHotEncoder()
embarked_train = ohe.fit_transform(train_df[["Embarked"]]).toarray()
embarked_train_df = pd.DataFrame(embarked_train, columns=ohe.get_feature_names_out(['Embarked'])) #df yaptım , makinenin oluşturduğu isimsiz sütunlara otomatik isim etiketi yapıştırmasını sağlar. son kod.
embarked_train_df.index = train_df.index #hata almamak için önemli,indexleri düzelt
train_df = train_df.drop("Embarked", axis=1) #eski embarked sütunun atıp
train_df = pd.concat([train_df, embarked_train_df], axis=1) # yerine yeni oluştudugumuz sütunu ekledik.

embarked_train_df
```

```python
train_df
```

```python
#------ Test Verisi İçin İşlemler (Aynısı) ---
embarked_test = ohe.transform(test_df[["Embarked"]]).toarray() #fit yok
embarked_test_df = pd.DataFrame(embarked_test, columns=ohe.get_feature_names_out(["Embarked"]))
embarked_test_df.index = test_df.index
test_df = test_df.drop("Embarked", axis=1)
test_df = pd.concat([test_df, embarked_test_df], axis=1)

embarked_test_df
```

```python
test_df
```

```python
y = train_df["Survived"]
X = train_df.drop(["Survived", "PassengerId"], axis=1)

from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=0)

from sklearn.preprocessing import StandardScaler #verilerin ölçeklenmesi yani genelde -3 ile 3 arası stanadart spamsı 1 olan değerler üretmek için
sc = StandardScaler()

X_train = sc.fit_transform(x_train)
X_test = sc.transform(x_test)

from sklearn.linear_model import LinearRegression
regressor = LinearRegression()
regressor.fit(x_train,y_train) #X_trainden y_traini öğren

y_pred = regressor.predict(X_test) # x in test olarak ayrılmış kısmını (X_test) tahmin eder
y_pred
X
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split

# Orijinal X verisinden sadece bu sütunları alıyoruz.
X_opt = X.iloc[:, [0, 1, 2, 3]].values 

# Optimize edilmiş veriyi tekrar eğitim ve test olarak ayırıyoruz.
X_train, X_test, y_train, y_test = train_test_split(X_opt, y, test_size=0.33, random_state=0)

regressor_final = LinearRegression()
regressor_final.fit(X_train, y_train)

y_pred_final = regressor_final.predict(X_test)

print("Final Tahminleri Oluşturuldu.")

y_pred_final[:10]
```

```python
from sklearn.metrics import accuracy_score

# Tahminleri 0 veya 1'e yuvarlıyoruz 
y_pred_rounded = [round(value) for value in y_pred_final]

# Doğruluk oranını hesaplıyoruz
dogruluk = accuracy_score(y_test, y_pred_rounded)

print(f"Modelin Doğruluk Oranı : %{dogruluk * 100}")
```
